Много пропусков в столбце 'Age', но выбросить эту фичу мы не можем, поэтому нужно заменить, например, медианой или средним. Но можно обратить внимание, что в именах присутмтвуют такие слова как Миссис или Мистер, что может сведетельствовать о возрасте. Поэтому можно извлеч из имен эту информацию и использовать для нахождения вохраста.

```python
data['Status']=data.Name.str.extract('([A-Za-z]+)\.')
status = list(data['Status'].unique())
median_age = list(data.groupby('Status')['Age'].median())
for i in range(len(status)):
    data.loc[(data.Age.isnull())&(data.Status==status[i]),'Age']=median_age[i]
data.isnull().sum()
```

**Summary**

PassengerId - обычное число, не дает никакой информациио данных

Survived - целевая метрика, оставляем

Pclass - важные данные, оставляем

Name - не несет нужной информации, удаляем

Sex - меняем значения 'male' на 0, 'female' на 1

Age - возраст, оставляем

SibSp и Pearch -  добавим одну фичу - размер семьи(Family_size)

Ticket - случайная строка, удаляем

Fare - оставляем

Cabin - очень разреженные данные, следует удалить столбец

Embarked - оставляем, кодируем

Status - оставляем, кодируем


```python
for i in range(len(status)):
    data.loc[(data.Status==status[i]),'Status']= i + 1
data['Sex'].replace(['male','female'],[0,1],inplace=True)
data['Embarked'].replace(['S','C','Q'],[0,1,2],inplace=True)
data['Family_size']=data['Parch']+data['SibSp']
data.drop(['PassengerId', 'Name', 'Ticket','Cabin'],axis=1,inplace=True)
```

```python
data['Status'] = data['Status'].astype(float)
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(clear_data, test_size=0.2)
```

```python
x, y = train.drop(columns=['Fare']), train['Fare']
x_t, y_t = test.drop(columns=['Fare']), test['Fare']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
def eval_model(x, y, x_t, y_t, model):
  fig = plt.gcf()
  fig.set_size_inches(18.5, 10.5)
  reg = model.fit(x, y)
  x_s = list(range(len(x)))
  plt.scatter(x_s, y - reg.predict(x))
  #plt.scatter(x_s, y)
  plt.title('Reg') 
  plt.xlabel('N sample') 
  plt.ylabel('Fare_ost') 
  plt.legend(['Fare_ost'])
  plt.show()
  print("Coefficient of determination(test): ", reg.score(x_t, y_t))
  print("Coefficient of determination(train): ", reg.score(x, y))
  print("MSE(test): ", mean_squared_error(y_t, reg.predict(x_t)))
  print("MSE(train): ", mean_squared_error(y, reg.predict(x)))
```

```python
eval_model(x, y, x_t, y_t, LinearRegression())
```

Значения стали намного лучше. Попробуем другие модели

```python
from sklearn.ensemble import GradientBoostingRegressor
eval_model(x, y, x_t, y_t, GradientBoostingRegressor())
```

Градиентый бустинг лучше справляется с поставленной задечей, как видно по коэф.детерминации и ошибке.

```python
from sklearn.ensemble import RandomForestRegressor
eval_model(x, y, x_t, y_t, RandomForestRegressor())
```
